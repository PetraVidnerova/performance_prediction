```python
import sys
import pandas as pd
```

```python
sys.path.append("HW-NAS-Bench")
```

```python
from hw_nas_bench_api import HWNASBenchAPI as HWAPI
hw_api = HWAPI("HW-NAS-Bench/HW-NAS-Bench-v1_0.pickle", search_space="nasbench201")
```

```python
df_list = []
for i in range(15625):
    HW_metrics = hw_api.query_by_index(i, "cifar10")
    network = hw_api.get_net_config(i, "cifar10")

    df_list.append(
        {
            "net": network["arch_str"],
            "eyeriss_energy": HW_metrics["eyeriss_energy"],
            "edgegpu_energy": HW_metrics["edgegpu_energy"],
            "fpga_energy": HW_metrics["fpga_energy"]
        }
    )
```

```python
df = pd.DataFrame(df_list).set_index("net")
df
```

```python
df.to_csv("mydata/hw_energy.csv")
```
